# Q-->3:How to Convert the index of a series into a column of a dataframe.



```python
import pandas as pd 
  
df = pd.DataFrame({'Roll Number': ['20ME21', '20ME22', '20ME33', '20ME44'], 
                   'Name': ['Shashank', 'Ram', 'Shivani', 'ali'], 
                   'Marks In Percentage': [97, 90, 99, 92], 
                   'Grade': ['A', 'A', 'C', 'B'], 
                   'Subject': ['production', 'automobile', 'biotechnology', 'electrical']}) 
  
# Printing the dataframe 
df['index'] = df.index 
df



```
